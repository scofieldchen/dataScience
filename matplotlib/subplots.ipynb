# 创建子图

一个图表对象(Figure)可以包含很多坐标(Axes)对象，单个坐标称为子图(subplot)。

maplotlib提供了3种创建子图的方式：

1. fig.add_axes
2. plt.subplots
3. plt.GridSpec

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use("ggplot")
```

## 1. fig.add_axes

1. 用plt.figure()创建图表对象，图表是坐标的容器。
2. 调用fig.add_axes()在图表的任意位置添加坐标，该方法接收一个包含4个数字的列表: $[x, y, width, height]$，分别代表子图的左下角的(x,y)，子图的宽度和高度，这四个数字的取值范围都是$[0,1]$，代表相对位置和大小。
3. 调用ax.plot或其它方法将数据映射到对应的坐标。

```python
fig = plt.figure()  # 创建图表

ax1 = fig.add_axes([0, 0.5, 0.45, 0.45])  # 在图表的左上角创建一个坐标/子图
ax2 = fig.add_axes([0.5, 0, 0.45, 0.45])  # 在图表的右下方创建一个坐标/子图

x = np.linspace(-10, 10, 100)

ax1.plot(x, np.sin(x), "-r")
ax2.plot(x, np.cos(x), "-or")
```

## 2. plt.subplots

plt.subplots是创建多个网格子图的接口,它返回包含两个元素的元组,第一个元素是Figure对象,第二个元素是坐标集合,存储在n维数组中.

```python
fig, axes = plt.subplots(nrows=2, ncols=3, sharex="col", sharey="row")

print(fig)
print(axes)  # 2*3矩阵

for i in range(2):
    for j in range(3):
        axes[i, j].text(0.5, 0.5, str((i, j)), ha="center", fontsize=15)
```

## 3. plt.GridSpec

创建可合并的子图,类似于excel中合并和拆分单元格的功能.

1. 调用plt.GridSpec(nrows, ncols)创建网格的'蓝图'.
2. 用切片的方式'合并'子图.
3. 调用ax.plot()将数据映射到图表.

```python
grid = plt.GridSpec(nrows=2, ncols=2)  # 创建2*2网格

ax1 = plt.subplot(grid[0, 0])  # 网格对象可索引和切片
ax2 = plt.subplot(grid[0, 1])
ax3 = plt.subplot(grid[1, :])  # 合并第一行所有列的子图

x = np.linspace(0, 10, 30)

ax1.plot(x, np.sin(x), "-r")
ax2.plot(x, np.cos(x), "-b")
ax3.plot(x, np.sin(x + 10), "-y", marker="o")
```
