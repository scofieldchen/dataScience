```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## 2. 简单线图

绘图步骤：

1. 准备数据
2. 创建图表(Figure)和坐标(Axes)对象，图表是坐标的容器
3. 调用坐标的绘图接口，如'plot'，将数据映射到图表
4. 调整样式，如颜色，坐标轴范围等

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

fig = plt.figure()
ax = plt.subplot()

ax.plot(x, y)  # 调用坐标对象的绘图接口，映射数据
ax.plot(x, y2)  # 多次调用继续添加数据到图表
```

## 3. 调整颜色和样式

* color参数控制颜色
* linestyle参数控制线条样式

```python
x = np.arange(100)

fig = plt.figure()
ax = plt.subplot()

ax.plot(x, x, color="red")  # 可以传递很多不同类型的参数，颜色名字是最方便的
ax.plot(x, x + 10, color="blue")
ax.plot(x, x + 20, color="yellow")
ax.plot(x, x + 30, color="black")
```

```python
x = np.arange(100)

fig = plt.figure()
ax = plt.subplot()

ax.plot(x, x, linestyle="solid")  # 实线，等价于'-'
ax.plot(x, x + 10, linestyle="dashed")  # 虚线，等价于'--'
ax.plot(x, x + 20, linestyle="dashdot")  # 点画线，等价于'-.'
ax.plot(x, x + 30, linestyle="dotted")  # 点线，等价于':'
```

## 4. 调整坐标轴的范围

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
ax = plt.subplot()

ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12)  # 等价于plt.xlim(-2, 12)
ax.set_ylim(-1.5, 1.5)  # 等价于plt.ylim(-1.5, 1.5)
```

## 5. 添加标签和图例

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
ax = plt.subplot()

ax.plot(x, np.sin(x), label="sin(x)")
ax.plot(x, np.cos(x), label="cos(x)")

ax.set_title("Simple Sin Function")
ax.set_xlabel("x")
ax.set_ylabel("y=f(x)")

ax.legend()
```

## 6. 高级接口ax.set()

```python
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(1,1)

ax.plot(x, np.sin(x), label="sin(x)")
ax.plot(x, np.cos(x), label="cos(x)")

ax.set(xlim=(-2, 12), ylim=(-1.5, 1.5),
       xlabel="x", ylabel="y=f(x)",
       title="Simple Line Plot")

ax.legend()
```
