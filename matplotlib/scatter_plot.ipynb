```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 1. 散点图

### 1.1 用ax.plot绘制散点图

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10, 6.18))

ax.plot(x, y, "o")  # 第3个参数是'fmt'，控制图案类型，'-'代表实线，'o'代表散点
```

### 1.2 用线将点连接在一起

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10, 6.18))

# ax.plot(x, y, "-o")  # fmt允许将多种类型的图案连接在一起
# ax.plot(x, y, marker="o")

# 通过调整plot参数调整图案的样式
ax.plot(
    x, y, "-o",  # 同时出现线和圆圈
    color="red",  # 同时控制线和圆圈边缘的颜色
    markersize=15,  # 控制点的大小
    linewidth=4,  # 控制线的大小
    markerfacecolor="blue",  # 控制圆圈填充的颜色
    markeredgecolor="blue",  # 控制圆圈边缘的颜色
    markeredgewidth=4  # 控制圆圈边缘的大小
)
```

### 1.3 散点类型

```python
markers = ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']

fig, ax = plt.subplots(figsize=(10, 6.18))

for marker in markers:
    ax.plot(np.random.rand(5), np.random.rand(5), marker, label=f"marker={marker}")
    
ax.legend()
```

## 2. ax.scatter

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10, 6.18))

ax.scatter(x, y, marker="o")
```

ax.scatter和ax.plot的区别在于，前者能够改变每一个点的样式，包括填充颜色，边缘颜色，大小等。

ax.scatter在绘制复杂的散点图时非常有用，假设我们想加入第三个维度，对数值变量x,y进行分类，这时可以将分类变量的水平值映射为不同颜色。

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)  # 传递给参数'c'将自动映射为不同的颜色
sizes = 1000*np.random.randn(100)  # 传递给参数's'将自动映射为不同的大小

fig, ax = plt.subplots(figsize=(10, 6.18))

ax.scatter(
    x, y,
    c=colors,  # 将随机数映射为一系列颜色
    s=sizes,  # 将随机数映射为一系列像素
    alpha=0.3,  # 调控颜色深浅
    cmap="viridis"  # 选择调色板
)
```
