```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use("ggplot")
```

```python
x = np.linspace(0, 10, 20)
dy = 0.5
y = np.sin(x) + dy * np.random.randn(20)

fig, ax = plt.subplots(figsize=(10, 6.18))

ax.errorbar(
    x=x, y=y, yerr=dy,
    fmt="o",
    color="black",
    ecolor="gray",
    elinewidth=3,
    capsize=0
)

ax.set(title="Simple Error Plot")
```
