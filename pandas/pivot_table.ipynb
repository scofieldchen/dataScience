## 数据透视表(Pivot Table)

数据透视表：从不同维度汇总数值变量，维度即分类变量，汇总方式包含求和，均值，方差等。

在pandas中用df.pivot_table()方法实现数据透视表。

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

使用数据透视表要考虑的要点：

1. 汇总哪个数值变量？
2. 根据哪些维度对数值变量进行分类？
3. 使用什么样的汇总方式？
4. 是否添加边缘汇总？

### 1.1 维度

根据性别和等级分类，计算存活率均值。默认汇总方式是计算均值。

```python
titanic.pivot_table(
    values="survived",
    index="sex",  # index和columns决定分割数据的维度
    columns="class"
)
```

多于两个维度。

```python
age = pd.cut(titanic["age"], bins=[0, 18, 80])  # 数值变量 ==> 分类变量

titanic.pivot_table(
    values="survived",
    index=["sex", age],
    columns=["class", "embark_town"]
)
```
