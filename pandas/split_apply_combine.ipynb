## 1. split-apply-combine

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

读取数据

```python
planets = sns.load_dataset("planets")
planets.head()
```

### 1.1 groupby

df.groupby()方法创建'分组数据框'对象，可理解为数据框的集合。

```python
grouped = planets.groupby("method")

type(grouped)
```

对于分组数据框，最常见的操作是计算分类数据的描述统计量。

```python
grouped.mean()
```

```python
grouped["distance"].describe()  # 索引方法依旧适用，选择某一个变量，然后计算常规统计量
```

'分组数据框'是可迭代对象，是(key, df)的集合。

```python
for g,df in grouped:
    print(f"group={g}, shape={df.shape}")
```

### 1.2 aggregate

```python
grouped.aggregate(["min", max, np.sum])
```

```python
grouped.aggregate({"number": np.min, "distance": np.max})
```

### 1.3 filter

```python
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': np.random.randint(0, 10, 6)},
                   columns = ['key', 'data1', 'data2'])
df
```

```python
df.groupby("key").filter(lambda x: x["data1"].sum() > 4)
```

### 1.4 transform

```python
df.groupby("key").transform(lambda x: x - x.mean())
```

### 1.5 apply

```python
def add_data(x):
    x["data3"] = x["data1"] + x["data2"].mean()
    return x

df.groupby("key").apply(add_data)
```
